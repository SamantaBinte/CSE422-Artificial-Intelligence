```python
import random

f=open("/content/Input1.txt", "r")

total_trans=f.readline()
total_trans=int(total_trans)


register=f.readlines()
reg=[]
for t in register:
  t=t.split()
  reg.append(t)
#print(reg) #taking all inputs and creating 2D list of the register




def fitness(register, population):
  fit_score=[]

  for each_chr in population:
    sum=0
    for i in range(len(each_chr)):
      if int(each_chr[i])==1:
        if register[i][0]=="l":
          sum-=int(register[i][1])
        else:
          sum+=int(register[i][1])
    #print(sum)
    fit_score.append(sum)
  return fit_score


def selection(population, fit_score):
  for i in range(len(population)):
    sort=sorted(fit_score)
    sort=sort[:-1] #eliminating the least fit
    parent_list=[]
    for i in sort:
      index=fit_score.index(i)
      parent_list.append(population[index])
    return parent_list

def crossover(parent_list):
  offspring_list=[]
  parent1=parent_list[0]
  parent2=parent_list[1]
  parent3=parent_list[2]
  crossover_point = random.randint(0, total_trans-1)
  offspring1 = parent1[:crossover_point] + parent2[crossover_point:]
  offspring2 = parent2[:crossover_point] + parent1[crossover_point:]
  offspring3 = parent2[:crossover_point] + parent3[crossover_point:]
  offspring4 = parent3[:crossover_point] + parent2[crossover_point:]

  offspring_list.append(offspring1)
  offspring_list.append(offspring2)
  offspring_list.append(offspring3)
  offspring_list.append(offspring4)
  return offspring_list


def mutation(parent1):
  mutation_point = random.randint(0, total_trans-1)
  if parent1[mutation_point]=="1":
    offspring1 = parent1[:mutation_point] +"0"+parent1[mutation_point+1:]
  else:
    offspring1 = parent1[:mutation_point] +"1"+parent1[mutation_point+1:]
  return offspring1


#creating a population
population=[]
for i in range(4):
  chromosome=''
  for j in range(total_trans):
    x=random.randint(0,1)
    chromosome+=str(x)
  #print(chromosome)
  population.append(chromosome)
#print(population)


#genetic algorithm

x=-1
most_fit=""
max_iteration=200

for i in range(max_iteration):

  #fitness_check
  fitness_score=fitness(reg, population) #a list containing all fitness score

  #goal_test
  if 0 in fitness_score:
    index=fitness_score.index(0)
    most_fit= population[index]
    break

  #parent selection
  parents=selection(population, fitness_score) #a list containing fittest choromosomes


  # Generate new offspring through crossover and mutation
  offspring_list1 = crossover(parents)
  for offspring in offspring_list1:
    offspring=mutation(offspring)

  #creating a new population
  population=offspring_list1


print(x)
print(most_fit)



```
